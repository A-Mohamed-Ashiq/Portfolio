```python
State_counts = gba_data['State'].value_counts()
```

```python
#Creating barchart
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

for index, value in enumerate(State_counts):
    plt.text(value, index, str(value), ha='left', va='center')
    
State_counts.plot(kind='barh', color='skyblue' )
plt.grid(True, alpha=0.2, color='gray')
plt.title('Number of times a State got Promoted by DMOs on Twitter')
plt.xlabel('Count')
plt.ylabel('States in India')

plt.show()
```

```python
gba_data['DateDay'].value_counts()
```

```python
DateDay_counts = gba_data['DateDay'].value_counts()

print(DateDay_counts)
```

```python
plt.figure(figsize=(10, 6))
  
DateDay_counts.plot(kind='bar' )
plt.grid(True, alpha=0.2, color='gray')
plt.title('Tweet Distribution by Day of the Week')
#plt.xlabel('Days')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Add count labels on each bar
for i, count in enumerate(DateDay_counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

```python
#Bar Plot for Vividness
plt.figure(figsize=(8, 6))
gba_data['Vividness'].value_counts().plot(kind='bar', color='purple')
plt.title('Vividness of Tweets')
plt.xlabel('Vividness')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for Time
plt.figure(figsize=(8, 6))
time_counts = gba_data['Time'].value_counts()
plt.pie(time_counts, labels=['Business Hours', 'Non-Business Hours'], autopct='%1.1f%%', colors=['lightblue', 'lightyellow'])
plt.title('Tweet Distribution by Time')
plt.show()
```

```python
# Grouping the data by OpnHours and calculating mean engagement metrics
engagement_metrics = gba_data.groupby('OpnHours').mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
bar_width = 0.2
opacity = 0.8

index = range(len(engagement_metrics))
plt.bar(index, engagement_metrics['retweet_count'], bar_width, alpha=opacity, color='b', label='Retweets')
plt.bar([i + bar_width for i in index], engagement_metrics['reply_count'], bar_width, alpha=opacity, color='g', label='Replies')
plt.bar([i + 2*bar_width for i in index], engagement_metrics['like_count'], bar_width, alpha=opacity, color='r', label='Likes')
plt.bar([i + 3*bar_width for i in index], engagement_metrics['quote_count'], bar_width, alpha=opacity, color='y', label='Quotes')

plt.xlabel('Time of Day')
plt.ylabel('Average Count')
plt.title('Engagement Metrics During Business vs. Non-Business Hours')
plt.xticks([i + 1.5 * bar_width for i in index], engagement_metrics['OpnHours'])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Grouping the data by OpnHours and ContentType and counting the occurrences
content_counts = gba_data.groupby(['OpnHours', 'ContentType']).size().unstack(fill_value=0)

# Plotting the bar plot
plt.figure(figsize=(10, 6))
bar_width = 0.35
opacity = 0.8

index = range(len(content_counts))
plt.bar(index, content_counts['Information'], bar_width, alpha=opacity, color='b', label='Information')
plt.bar([i + bar_width for i in index], content_counts['Interaction'], bar_width, alpha=opacity, color='g', label='Interaction')

plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.title('Content Type Distribution During Business vs. Non-Business Hours')
plt.xticks([i + bar_width / 2 for i in index], content_counts.index)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Grouping the data by Time and Corrected_Sentiment_Flag and counting the occurrences
sentiment_counts = gba_data.groupby(['Time', 'Corrected_Sentiment_Flag']).size().unstack(fill_value=0)

# Plotting the bar plot
plt.figure(figsize=(8, 6))

bar_width = 0.35
opacity = 0.8

index = range(len(sentiment_counts))
plt.bar(index, sentiment_counts[1], bar_width, alpha=opacity, color='b', label='1')
plt.bar([i + bar_width for i in index], sentiment_counts[0], bar_width, alpha=opacity, color='r', label='0')

plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.title('Sentiment Flag Distribution by Time')
plt.xticks([i + bar_width / 2 for i in index], sentiment_counts.index)
plt.legend()

plt.tight_layout()
plt.show()
```
