```python
import pandas as pd

gba_data=pd.read_csv("Data LIWC 01 02 23.csv")
```

```python
display(gba_data)
```

```python
# Calculate corrected sentiment score
gba_data['Corrected_Sentiment'] = gba_data['Positive'] + gba_data['Negative']
```

```python
gba_data['Corrected_Sentiment'].value_counts()
```

```python
# Define a function to categorize sentiment
def categorize_sentiment(score):
    if score <= 0:
        return '0'
   
    else:
        return '1'

# Apply the function to create a new column 
gba_data['Corrected_Sentiment_Flag'] = gba_data['Corrected_Sentiment'].apply(categorize_sentiment)
```

```python
display(gba_data)
```

```python
gba_data['Corrected_Sentiment_Flag'].value_counts()
```

```python
Total_Sentiment_Flag_counts = gba_data['Corrected_Sentiment_Flag'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10, 6))
  
Total_Sentiment_Flag_counts.plot(kind='bar' )
plt.grid(True, alpha=0.2, color='gray')
plt.title('Distribution of Sentiments')
#plt.xlabel('Days')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Add count labels on each bar
for i, count in enumerate(Total_Sentiment_Flag_counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

```python
# Sort the value counts
sorted_counts = gba_data['Corrected_Sentiment'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(8, 6))
bar_plot = sorted_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.grid(True, alpha=0.2, color='gray')
plt.title('Distribution of Corrected Sentiment Scores')
plt.xlabel('Corrected Sentiment Score')
plt.ylabel('Count')

for i, count in enumerate(sorted_counts):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.show()
```

```python
# Save DataFrame to CSV file
gba_data.to_csv('Data LIWC 01 02 23.csv', index=False)
```
